```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv("Desktop/IPL_ML/matches.csv")
delivery = pd.read_csv("Desktop/IPL_ML/deliveries.csv")
```

```python
matches.shape
```

```python
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
total_score_df.shape
```

```python
#First inning Score 
total_score_df = total_score_df[total_score_df["inning"]==1]
```

```python
#Merging this first inning score into the matches dataframe
match_df = matches.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
#Remove the IPL Teams who are not part of IPL 2025
match_df['team1'].unique()
```

```python
teams = [
          'Mumbai Indians', 
         'Kolkata Knight Riders',
         'Rajasthan Royals',  
         'Chennai Super Kings',
        'Sunrisers Hyderabad',
        'Delhi Capitals', 
         'Punjab Kings',
         'Lucknow Super Giants', 
         'Gujarat Titans',
         'Royal Challengers Bengaluru']
```

```python
match_df['team1'] = match_df['team1'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')
match_df['team2'] = match_df['team2'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')

match_df['team1'] = match_df['team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['team2'] = match_df['team2'].str.replace('Kings XI Punjab','Punjab Kings')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
match_df
```

```python
match_df.team1.unique()
```

```python
match_df.head()
```

```python
#We need matchid,city,winner,total_runs
match_df = match_df[['id','city','winner','total_runs']]
match_df
```

```python
#Joining above match df with delivery
delivery_df = match_df.merge(delivery, left_on='id', right_on='match_id')
```

```python
#I want only of 2nd innings
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.shape
```

```python
# First, ensure the column is numeric
delivery_df['total_runs_y'] = pd.to_numeric(delivery_df['total_runs_y'], errors='coerce')

# Now apply groupby and cumsum
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df['current_score']
```

```python
delivery_df
```

```python
#Now we calculate the runs left
delivery_df['runs_left'] = delivery_df['total_runs_x']+1 - delivery_df['current_score']
```

```python
delivery_df
```

```python
#Getting how many balls are left  
delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
delivery_df
```

```python
# Step 1: Fix player_dismissed column (again for safety)
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: "1" if x != "0" else "0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)

# Step 2: Calculate cumulative wickets and cap them at 10
delivery_df['wickets_lost'] = (
    delivery_df.groupby('id')['player_dismissed']
    .cumsum()
    .clip(upper=10)  # cap max wickets at 10
)

# Step 3: Remaining wickets
delivery_df['wickets_left'] = 10 - delivery_df['wickets_lost']
```

```python
delivery_df.head()
```

```python
#Current Run rate 
#Runs / Over
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
#Require Run rate
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
def winner(team):
    return 1 if team['batting_team'] == team['winner'] else 0
```

```python
delivery_df['results'] = delivery_df.apply(winner,axis=1)
delivery_df
```

```python
required_data = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','results']]
required_data
```

```python
final_data = required_data.sample(required_data.shape[0])
```

```python
final_data
```

```python
final_data.dropna(inplace=True)
```

```python
final_data = final_data[final_data['balls_left']!=0]
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x = final_data.iloc[:, :-1]
y = final_data.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.25, random_state=100)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# Defining a transformer for one-hot encoding categorical features
trf = ColumnTransformer([
    ('trf', OneHotEncoder(drop='first', sparse_output=False), ['batting_team', 'bowling_team', 'city'])
],
remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
# Creating a pipeline with column transformation and logistic regression
pipe = Pipeline([
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
# Fitting the pipeline to the training data
pipe.fit(x_train, y_train)
```

```python
# Making predictions on the test data
y_pred = pipe.predict(x_test)
```

```python
# Calculating the accuracy of the model
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Predicting class probabilities for the first instance in the test set
pipe.predict_proba(x_test)[0]
```

```python
# Saving the trained pipeline as a binary file
import pickle
pickle.dump(pipe, open('pipe.pkl', 'wb'))
```
